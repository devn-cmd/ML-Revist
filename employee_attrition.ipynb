```python
income_99=df.MonthlyIncome.quantile(0.99)
income_01 = df.MonthlyIncome.quantile(0.01)

outlier_count=((df.MonthlyIncome > income_99) | (df.MonthlyIncome < income_01)).sum()
df.MonthlyIncome=df.MonthlyIncome.clip(upper=income_99,lower=income_01)
print(f"\n✅ [MonthlyIncome] Capped {outlier_count} outlier(s) at 99th percentile (${income_99:,.0f})")
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.2,random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
x_train=scaler.fit_transform(x_train)
x_test=scaler.transform(x_test)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier,GradientBoostingClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier     
import warnings
warnings.filterwarnings('ignore')
from sklearn.metrics import (accuracy_score, precision_score, recall_score,
                              f1_score, confusion_matrix, classification_report)
from sklearn.model_selection import cross_val_score,GridSearchCV,RandomizedSearchCV

```

```python
models = {
    'Logistic Regression'    : LogisticRegression(random_state=42),
    'Decision Tree'          : DecisionTreeClassifier(random_state=42),
    'Random Forest'          : RandomForestClassifier(n_estimators=100, random_state=42),
    'Gradient Boosting'      : GradientBoostingClassifier(random_state=42),
    'SVM'                    : SVC(random_state=42),
    'KNN'                    : KNeighborsClassifier(n_neighbors=5),
}
```

```python
# hyperparameter tuning

param_grids = {
    'Logistic Regression': {
        'search' : 'grid',
        'params' : {
            'C'       : [0.01, 0.1, 1, 10, 100],
            'penalty' : ['l1', 'l2'],
            'solver'  : ['liblinear', 'saga'],
        }
    },
    'Decision Tree': {
        'search' : 'grid',
        'params' : {
            'max_depth'        : [3, 5, 7, 10, None],
            'min_samples_split': [2, 5, 10],
            'criterion'        : ['gini', 'entropy'],
        }
    },
    'Random Forest': {
        'search' : 'random',   # RandomizedSearch — too many combos
        'params' : {
            'n_estimators'     : [50, 100, 200, 300],
            'max_depth'        : [3, 5, 10, None],
            'min_samples_split': [2, 5, 10],
            'max_features'     : ['sqrt', 'log2'],
        }
    },
    'Gradient Boosting': {
        'search' : 'random',   # RandomizedSearch — too many combos
        'params' : {
            'n_estimators' : [50, 100, 200],
            'learning_rate': [0.01, 0.05, 0.1, 0.2],
            'max_depth'    : [2, 3, 4, 5],
            'subsample'    : [0.6, 0.8, 1.0],
        }
    },
    'SVM': {
        'search' : 'grid',
        'params' : {
            'C'     : [0.1, 1, 10, 100],
            'kernel': ['linear', 'rbf', 'poly'],
            'gamma' : ['scale', 'auto'],
        }
    },
    'KNN': {
        'search' : 'grid',
        'params' : {
            'n_neighbors': [3, 5, 7, 9, 11, 15],
            'metric'     : ['euclidean', 'manhattan'],
            'weights'    : ['uniform', 'distance'],
        }
    },
}

results = []

for name, model in models.items():

    # ── Hyperparameter Tuning ─────────────────────────────────
    grid_info = param_grids[name]

    if grid_info['search'] == 'grid':
        search = GridSearchCV(
            model, grid_info['params'],
            cv=5, scoring='f1', n_jobs=-1
        )
    else:
        search = RandomizedSearchCV(
            model, grid_info['params'],
            n_iter=30, cv=5, scoring='f1',
            n_jobs=-1, random_state=42
        )

    search.fit(x_train, y_train)         # fit with tuning
    best_model = search.best_estimator_  # get best version

    # ── Predict with Best Model ───────────────────────────────
    y_pred = best_model.predict(x_test)

    # ── Metrics ───────────────────────────────────────────────
    acc  = accuracy_score(y_test, y_pred)
    prec = precision_score(y_test, y_pred, zero_division=0)
    rec  = recall_score(y_test, y_pred, zero_division=0)
    f1   = f1_score(y_test, y_pred, zero_division=0)

    # ── Cross Validation on Best Model ────────────────────────
    # cv_scores = cross_val_score(best_model, x_train, y_train, cv=5, scoring='f1')
    # cv_mean   = cv_scores.mean()

    results.append({
        'Model'      : name,
        'Best Params': search.best_params_,  # ← shows what tuning found
        'Accuracy'   : round(acc,     4),
        'Precision'  : round(prec,    4),
        'Recall'     : round(rec,     4),
        'F1 Score'   : round(f1,      4),
        'CV F1'      : round(search.best_score_, 4),
    })

    print(f"\n {name}")
    print(f"   Best Params : {search.best_params_}")
    print(f"   Accuracy    : {acc:.4f}  |  F1 Score : {f1:.4f}")
    print(f"   Precision   : {prec:.4f}  |  Recall   : {rec:.4f}")
    print(f"   CV F1       : {search.best_score_:.4f}")
```
