```python
x=df.select_dtypes(include=np.number)
x
```

```python
print("NaN count:", np.isnan(x).sum())
print("Inf count:", np.isinf(x).sum())
```

```python
# clustering
from sklearn.cluster import KMeans,AgglomerativeClustering,DBSCAN
from sklearn.metrics import silhouette_score, davies_bouldin_score
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
print("  ELBOW METHOD — Finding Optimal K")
inertias=[]
silhouette_scores=[]
k_range=range(2,11)

for k in k_range:
    kmeans=KMeans(n_clusters=k,random_state=42,init='k-means++')
    kmeans.fit(x)
    inertias.append(kmeans.inertia_)
    sil=silhouette_score(x, kmeans.labels_)
    silhouette_scores.append(sil)
    print(f"   K={k}  |  Inertia: {kmeans.inertia_:>12.2f}  |  "
          f"Silhouette: {sil:.4f}")
    
```

```python

import matplotlib.pyplot as plt
plt.plot(range(2,11),inertias,marker='o')
plt.xlabel("n_clusters")
plt.ylabel('variance')
plt.title('elbow point')
```

```python
best_k = k_range[silhouette_scores.index(max(silhouette_scores))]
best_k
```

```python
kmeans=KMeans(n_clusters=best_k,random_state=42,init="k-means++")
kmeans_labels=kmeans.fit_predict(x)
kmeans_sil=silhouette_score(x, kmeans_labels)
kmeans_db=davies_bouldin_score(x, kmeans_labels)
```

```python
print(f"   Silhouette Score    : {kmeans_sil:.4f}  "
      f"(higher = better, max=1.0)")
print(f"   Davies-Bouldin Score: {kmeans_db:.4f}  "
      f"(lower = better, min=0.0)")
```

```python
df['KMeans_Cluster'] = kmeans_labels
df['KMeans_Cluster'] 
```

```python
print(f"\n📊 Cluster Profiles:")
cluster_profile = df.groupby('KMeans_Cluster').agg({
    'JobTitle'       : lambda x: x.mode()[0],
    'Salary_USD'     : 'mean',
    'YearsExperience': 'mean',
    'JobDemandScore' : 'mean',
    'Location'       : lambda x: x.mode()[0],
}).round(2)
cluster_profile.columns = ['Top JobTitle','Avg Salary','Avg Experience',
                            'Avg Demand','Top Location']
print(cluster_profile.to_string())
```

```python
print("\n" + "─" * 62)
print(f"  2. Hierarchical (Agglomerative) Clustering (K={best_k})")
print("─" * 62)
 
agg          = AgglomerativeClustering(n_clusters=best_k, linkage='ward')
agg_labels   = agg.fit_predict(x)
agg_sil      = silhouette_score(x, agg_labels)
agg_db       = davies_bouldin_score(x, agg_labels)
 
print(f"   Silhouette Score    : {agg_sil:.4f}")
print(f"   Davies-Bouldin Score: {agg_db:.4f}")
 
df['Hierarchical_Cluster'] = agg_labels
```

```python
dbscan       = DBSCAN(eps=2.5, min_samples=5)
dbscan_labels= dbscan.fit_predict(x)
n_clusters   = len(set(dbscan_labels)) - (1 if -1 in dbscan_labels else 0)
noise_points = (dbscan_labels == -1).sum()
 
print(f"   Clusters found      : {n_clusters}")
print(f"   Noise points (-1)   : {noise_points}")
 
if n_clusters > 1:
    # Silhouette only works if more than 1 cluster found
    mask     = dbscan_labels != -1
    dbscan_sil = silhouette_score(x[mask], dbscan_labels[mask])
    print(f"   Silhouette Score    : {dbscan_sil:.4f}")
else:
    dbscan_sil = 0
    print(f"   Silhouette Score    : N/A (only 1 cluster found)")
 
df['DBSCAN_Cluster'] = dbscan_labels
```

```python
print("  PCA — Dimensionality Reduction")
```

```python

pca        = PCA(n_components=2, random_state=42)
X_pca      = pca.fit_transform(x)
explained  = pca.explained_variance_ratio_
print(f"   Component 1 explains: {explained[0]*100:.2f}% of variance")
print(f"   Component 2 explains: {explained[1]*100:.2f}% of variance")
print(f"   Total explained     : {sum(explained)*100:.2f}% of variance")
```

```python
pca_kmeans       = KMeans(n_clusters=best_k, random_state=42, n_init=10)
pca_labels       = pca_kmeans.fit_predict(X_pca)
pca_sil          = silhouette_score(X_pca, pca_labels)
print(f"\n   KMeans on PCA data:")
print(f"   Silhouette Score    : {pca_sil:.4f}")
 
```

```python
df['PCA_Cluster']  = pca_labels
df['PCA_Component1'] = X_pca[:, 0]
df['PCA_Component2'] = X_pca[:, 1]
 
```

```python
print("\n" + "=" * 62)
print("  MODEL COMPARISON")
print("=" * 62)
 
comparison = pd.DataFrame([
    {'Model'             : 'KMeans',
     'Clusters'          : best_k,
     'Silhouette Score'  : round(kmeans_sil, 4),
     'Davies-Bouldin'    : round(kmeans_db,  4),
     'Notes'             : 'Need to specify K'},
    {'Model'             : 'Hierarchical',
     'Clusters'          : best_k,
     'Silhouette Score'  : round(agg_sil,    4),
     'Davies-Bouldin'    : round(agg_db,     4),
     'Notes'             : 'No retraining needed'},
    {'Model'             : 'DBSCAN',
     'Clusters'          : n_clusters,
     'Silhouette Score'  : round(dbscan_sil, 4),
     'Davies-Bouldin'    : 'N/A',
     'Notes'             : 'Finds K automatically'},
    {'Model'             : 'KMeans + PCA',
     'Clusters'          : best_k,
     'Silhouette Score'  : round(pca_sil,    4),
     'Davies-Bouldin'    : 'N/A',
     'Notes'             : 'Reduced to 2D for visualization'},
])
print(comparison.to_string(index=False))
```

```python
best_model = comparison.loc[
    comparison['Silhouette Score'].idxmax(), 'Model']
print(f"\n Best Model : {best_model}")
print(f"   (highest Silhouette Score)")
 
```
